# **1. Introduction**

## **1.i) Background**

Metal casting is one of the oldest and most widely used manufacturing processes for producing complex engineering components. The quality and performance of cast products are strongly influenced by the solidification process, which governs the thermal history, microstructure formation, defect generation, and final mechanical properties of the material. Among various casting materials, gray cast iron remains one of the most extensively used ferrous alloys due to its excellent castability, good machinability, high compressive strength, wear resistance, vibration damping capacity, and relatively low production cost.

The solidification of gray cast iron involves complex heat transfer and phase transformation phenomena. During cooling, the molten metal loses heat to the surrounding mold through conduction and convection mechanisms. As the temperature decreases from the pouring temperature to the liquidus temperature, sensible heat is removed from the liquid metal. Subsequently, latent heat is released during the liquid-to-solid phase transformation, followed by further sensible heat removal as the solidified casting cools to ambient temperature. The rate at which heat is extracted significantly influences the cooling rate, thermal gradient, solidification time, and solidification front movement, all of which affect the final casting characteristics.

---

## **1.ii) Heat Transfer During Solidification**

The solidification process can be described as a transient heat transfer problem involving energy transport within the casting and heat dissipation through the mold. Heat extraction occurs in three distinct stages:

1. **Liquid Cooling Stage:** Removal of sensible heat from the molten metal as it cools from the pouring temperature to the liquidus temperature.

2. **Phase Transformation Stage:** Release of latent heat during the transformation of liquid metal into solid metal within the mushy zone.

3. **Solid Cooling Stage:** Removal of sensible heat from the solidified casting until it reaches the ambient temperature.

The interaction between these stages determines the overall thermal behavior of the casting. Important parameters such as cooling rate, thermal diffusivity, temperature gradient, and solidification front velocity play critical roles in controlling the solidification process and the resulting material properties.

---

## **1.iii) Importance of Computational Modelling**

Experimental investigation of casting solidification is often expensive, time-consuming, and difficult to perform due to the high temperatures involved and the limited accessibility of internal temperature measurements. Computational modeling provides an efficient alternative by enabling the prediction of thermal behavior, phase transformation kinetics, and energy transfer mechanisms under various operating conditions.

Numerical methods allow researchers and engineers to simulate casting processes, evaluate the influence of material properties and process parameters, and optimize manufacturing conditions before conducting physical experiments. Among the available numerical techniques, the Finite Difference Method (FDM) offers a relatively simple and effective approach for solving transient heat conduction problems in one-dimensional domains.

---

## **1.iv) Gray Cast Iron Solidification**

Thermal diffusivity indicates the ability of a material to conduct thermal energy relative to its heat storage capacity, while Fourier and Biot numbers provide insight into transient heat conduction and surface heat transfer characteristics.

The understanding of these parameters has enabled the development of numerical models capable of simulating temperature evolution and thermal gradients during solidification.

# **5. Assumptions**

To simplify the numerical simulation and focus on the fundamental thermal behavior of gray cast iron during solidification, the following assumptions were made in the present study:

### **I. One-Dimensional Heat Transfer**

Heat transfer within the casting is assumed to occur only along the thickness direction. Temperature variations in the other two spatial directions are neglected, resulting in a one-dimensional transient heat conduction model.

### **II. Homogeneous Material Properties**

The gray cast iron casting is assumed to be homogeneous and isotropic throughout the simulation domain. Material properties are considered uniformly distributed within the casting.

### **III. Constant Thermophysical Properties**

The density, thermal conductivity, specific heat capacity, and latent heat of fusion of gray cast iron are assumed to remain constant during the entire cooling and solidification process. Temperature-dependent variations of material properties are neglected.

### **IV. Uniform Initial Temperature**

The molten gray cast iron is assumed to have a uniform initial temperature equal to the pouring temperature throughout the casting at the beginning of the simulation.

### **V. Uniform Mold Temperature**

The sand mold is assumed to be initially at a uniform room temperature before the pouring process begins.

### **VI. Pure Heat Conduction Within the Casting**

Heat transfer inside the molten and solidifying metal is assumed to occur only through conduction. Natural convection and fluid flow effects within the liquid metal are neglected.

### **VII. Negligible Radiation Heat Transfer**

Thermal radiation effects between the casting and the surroundings are assumed to be negligible compared to conductive and convective heat transfer mechanisms.

### **VIII. Constant Heat Transfer Coefficient**

The heat transfer coefficient at the casting–mold interface is assumed to remain constant throughout the cooling process. Changes due to air-gap formation or mold expansion are neglected.

### **IX. Linear Solid Fraction Evolution**

Within the mushy zone, the solid fraction is assumed to vary linearly between the liquidus temperature and the solidus temperature. This simplifies the representation of the phase transformation process.

### **X. Negligible Volume Change During Solidification**

The effects of solidification shrinkage, thermal contraction, and density changes due to phase transformation are neglected.

### **XI. No Internal Heat Generation**

No internal heat sources are present within the casting during cooling and solidification. The only energy exchange considered is heat removal to the surrounding mold and environment.

### **XII. Sharp Solidification Criteria**

The start and completion of solidification are determined based on the liquidus and solidus temperatures, respectively, without considering complex microstructural transformations.

### **XIII. Fixed Geometric Dimensions**

The dimensions of the casting remain constant during the simulation. Geometrical deformation due to thermal stresses or shrinkage is not considered.

### **XIV. Validity of Fourier Heat Conduction Theory**

The transient heat transfer process is assumed to obey Fourier's law of heat conduction throughout the simulation period.

### **XV. Numerical Stability and Grid Independence**

The selected spatial and temporal discretization parameters are assumed to satisfy the numerical stability requirements of the Finite Difference Method and provide sufficiently accurate results for engineering analysis.

---

## **Justification of Assumptions**

These assumptions are commonly employed in preliminary thermal and solidification analyses to reduce computational complexity while retaining the essential physics of heat transfer and phase transformation. Although certain real-world phenomena such as fluid flow, shrinkage, temperature-dependent properties, and multidimensional effects are neglected, the model provides a reasonable approximation of the solidification behavior of gray cast iron and is suitable for educational, research, and preliminary engineering investigations.

# **9. Python Implementation**

````python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. INPUT PARAMETERS & PHYSICAL CONSTANTS
# ==========================================

# Material Properties (Gray Cast Iron)
rho = 7200.0          # Density (kg/m^3)
k_l = 40.0            # Thermal conductivity - liquid (W/m*K)
k_s = 52.0            # Thermal conductivity - solid (W/m*K)
cp_l = 750.0          # Specific heat - liquid (J/kg*K)
cp_s = 550.0          # Specific heat - solid (J/kg*K)
L_fusion = 250000.0   # Latent heat of fusion (J/kg)

# Solidification Temperatures (Celsius)
T_pour = 1450.0       # Pouring temperature
T_L = 1200.0          # Liquidus temperature
T_S = 1150.0          # Solidus temperature
T_room = 30.0         # Ambient/Room temperature

# Casting Geometry
thickness = 0.02      # Half-thickness for symmetric 1D domain (m)
area = 0.01           # Cross-sectional area (m^2)
volume = thickness * area
mass = rho * volume   # Parameter 8: Mass of casting

# Mold Properties (Sand Mold)
k_mold = 0.8          # Thermal conductivity (W/m*K)
rho_mold = 1600.0     # Density (kg/m^3)
cp_mold = 900.0       # Specific heat (J/kg*K)

# ==========================================
# 2. NUMERICAL DISCRETIZATION SETUP
# ==========================================
Nx = 21               # Ideal number of nodes for spatial resolution
dx = thickness / (Nx - 1)
x = np.linspace(0, thickness, Nx)

# Strict explicit stability check (Fourier number safely set to 0.15)
alpha_max = k_s / (rho * cp_s)
dt = 0.15 * (dx**2) / alpha_max
t_max = 600.0                     # Total simulation time (s)
Nt = int(t_max / dt)

# Step-skipping factor for history tracking to keep processing light
sample_interval = max(1, Nt // 1000)

# Initialize Arrays
T = np.full(Nx, T_pour)
T_new = np.copy(T)
fs = np.zeros(Nx)
fs_old = np.zeros(Nx)

# Storage lists for parameter tracking over time
time_history = []
T_center_history = []
T_surface_history = []
fs_avg_history = []
cooling_rate_history = []
sensible_heat_history = []
latent_heat_history = []
heat_removed_history = []
alpha_avg_history = []
Fo_history = []
Bi_history = []
grad_surface_history = []
front_velocity_history = []

# Tracker variables for key metrics
solidification_time = None
max_cooling_rate = 0.0
front_idx = 0

# ==========================================
# 3. TRANSIENT SIMULATION LOOP
# ==========================================
print(f"Running corrected simulation over {Nt} timesteps...")

for istep in range(Nt):
    t = istep * dt

    # Vectorized solid fraction estimation
    fs = np.clip((T_L - T) / (T_L - T_S), 0.0, 1.0)

    # Vectorized dynamic material properties mapping
    k_node = k_l + fs * (k_s - k_l)
    cp_node = cp_l + fs * (cp_s - cp_l)

    # Capped transient Heat Transfer Coefficient to avoid t=0 singularity
    if t > 0.1:
        h = np.sqrt(k_mold * rho_mold * cp_mold / (np.pi * t))
    else:
        h = 2500.0  # Capped realistic initial contact HTC

    # --- Core FDM Solver Update Equations ---

    # A. Internal Nodes (Central Difference)
    k_plus = 0.5 * (k_node[2:] + k_node[1:-1])
    k_minus = 0.5 * (k_node[1:-1] + k_node[:-2])
    conduction = (k_plus * (T[2:] - T[1:-1]) - k_minus * (T[1:-1] - T[:-2])) / (dx**2)
    latent_source = rho * L_fusion * (fs[1:-1] - fs_old[1:-1]) / dt
    T_new[1:-1] = T[1:-1] + (dt / (rho * cp_node[1:-1])) * (conduction + latent_source)

    # B. Center Boundary Node (Symmetry: dT/dx = 0)
    k_p_center = 0.5 * (k_node[1] + k_node[0])
    conduction_center = (2.0 * k_p_center * (T[1] - T[0])) / (dx**2)
    latent_source_center = rho * L_fusion * (fs[0] - fs_old[0]) / dt
    T_new[0] = T[0] + (dt / (rho * cp_node[0])) * (conduction_center + latent_source_center)

    # C. Surface Boundary Node (Conduction balanced with Sand Mold Convection)
    k_m_surf = 0.5 * (k_node[-1] + k_node[-2])
    conduction_surf = (2.0 * k_m_surf * (T[-2] - T[-1])) / (dx**2)
    convection_surf = (2.0 * h * (T_room - T[-1])) / dx
    latent_source_surf = rho * L_fusion * (fs[-1] - fs_old[-1]) / dt
    # Corrected Division Factor: Applied uniformly to total thermal energy change
    T_new[-1] = T[-1] + (dt / (rho * cp_node[-1])) * (conduction_surf + convection_surf + latent_source_surf)

    # Track maximum cooling rate at center
    if istep > 0:
        cr = (T_center_history[-1] - T_new[0]) / dt if len(T_center_history) > 0 else 0.0
        if cr > max_cooling_rate and t > 0.5:
            max_cooling_rate = cr

    # Track global solidification completion time
    avg_fs = np.mean(fs)
    if solidification_time is None and avg_fs >= 0.99:
        solidification_time = t

    # --- Periodic Downsampled Performance Tracking ---
    if istep % sample_interval == 0 or istep == Nt - 1:
        time_history.append(t)
        T_center_history.append(T_new[0])
        T_surface_history.append(T_new[-1])
        fs_avg_history.append(avg_fs)
        cooling_rate_history.append((T_center_history[-2] - T_new[0])/dt if len(T_center_history) > 1 else 0.0)

        sensible_energy = np.sum(rho * cp_node * (T_pour - T_new) * (thickness / Nx) * area)
        sensible_heat_history.append(sensible_energy)

        latent_energy = np.sum(rho * L_fusion * fs * (thickness / Nx) * area)
        latent_heat_history.append(latent_energy)

        heat_removed_history.append(sensible_energy + latent_energy)

        alpha_avg = np.mean(k_node / (rho * cp_node))
        alpha_avg_history.append(alpha_avg)

        Fo_history.append((alpha_avg * t) / (thickness**2))
        Bi_history.append((h * thickness) / np.mean(k_node))
        grad_surface_history.append((T_new[-2] - T_new[-1]) / dx)

        current_front_idx = np.where(fs >= 1.0)[0][-1] if np.any(fs >= 1.0) else 0
        if current_front_idx > front_idx:
            v_front = ((current_front_idx - front_idx) * dx) / (dt * sample_interval)
            front_idx = current_front_idx
        else:
            v_front = 0.0
        front_velocity_history.append(v_front)

    # Update states
    T = np.copy(T_new)
    fs_old = np.copy(fs)

if solidification_time is None:
    solidification_time = t_max

print("Simulation finished successfully with clean numeric bounds.\n")

# ==========================================
# 4. PRINT SUMMARY REPORT
# ==========================================
print("=========================================================")
print("            SOLIDIFICATION SIMULATION REPORT             ")
print("=========================================================")
print(f"8. Mass of Casting                : {mass:.4f} kg")
print(f"12. Total Solidification Time    : {solidification_time:.2f} seconds")
print(f"13. Maximum Cooling Rate         : {max_cooling_rate:.4f} °C/s")
print(f"6. Final Thermal Heat Removed    : {heat_removed_history[-1]/1000:.2f} kJ")
print(f"9. Total Sensible Heat Released  : {sensible_heat_history[-1]/1000:.2f} kJ")
print(f"5. Total Latent Heat Released    : {latent_heat_history[-1]/1000:.2f} kJ")
print(f"7. Mean Thermal Diffusivity      : {np.mean(alpha_avg_history):.4e} m^2/s")
print(f"11. Average Biot Number (Bi)     : {np.mean(Bi_history):.4f}")
print("=========================================================\n")

# ==========================================
# 5. VISUALIZATION AND DATA PLOTTING
# ==========================================
plt.figure(figsize=(14, 16))

plt.subplot(4, 2, 1)
plt.plot(time_history, T_center_history, label="Center Node", color='firebrick')
plt.plot(time_history, T_surface_history, label="Surface Node", color='darkorange')
plt.axhline(T_L, color='gray', linestyle='--', label='Liquidus')
plt.axhline(T_S, color='black', linestyle='--', label='Solidus')
plt.title("1. Temperature vs Time")
plt.xlabel("Time (s)"); plt.ylabel("Temperature (°C)"); plt.grid(True); plt.legend()

plt.subplot(4, 2, 2)
plt.plot(time_history, fs_avg_history, color='darkgreen')
plt.title("2. Average Solid Fraction vs Time")
plt.xlabel("Time (s)"); plt.ylabel("Solid Fraction"); plt.grid(True)

plt.subplot(4, 2, 3)
plt.plot(time_history, cooling_rate_history, color='blue')
plt.title("3. Cooling Rate (Center) vs Time")
plt.xlabel("Time (s)"); plt.ylabel("Cooling Rate (°C/s)"); plt.grid(True)

plt.subplot(4, 2, 4)
plt.plot(time_history, np.array(sensible_heat_history)/1000, label="Sensible", color='crimson')
plt.plot(time_history, np.array(latent_heat_history)/1000, label="Latent", color='gold')
plt.plot(time_history, np.array(heat_removed_history)/1000, label="Total Heat Removed", color='black', linestyle='-.')
plt.title("4 & 5. Heat Release Curves")
plt.xlabel("Time (s)"); plt.ylabel("Energy (kJ)"); plt.grid(True); plt.legend()

plt.subplot(4, 2, 5)
plt.plot(time_history, alpha_avg_history, color='purple')
plt.title("7. Thermal Diffusivity vs Time")
plt.xlabel("Time (s)"); plt.ylabel("Diffusivity (m^2/s)"); plt.grid(True)

plt.subplot(4, 2, 6)
plt.plot(time_history, Fo_history, color='teal')
plt.title("10. Fourier Number vs Time")
plt.xlabel("Time (s)"); plt.ylabel("Fourier Number"); plt.grid(True)

plt.subplot(4, 2, 7)
plt.plot(time_history, Bi_history, color='chocolate')
plt.title("11. Biot Number vs Time")
plt.xlabel("Time (s)"); plt.ylabel("Biot Number"); plt.grid(True)

plt.subplot(4, 2, 8)
plt.plot(time_history, grad_surface_history, color='magenta')
plt.title("14. Surface Temperature Gradient")
plt.xlabel("Time (s)"); plt.ylabel("Gradient (°C/m)"); plt.grid(True)

plt.tight_layout()
plt.show()
````
